```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(data=tips, x= 'sex', y='total_bill', estimator='mean')
```

```python
sns.countplot(data=tips, x='sex')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', split=True)
```

```python
sns.stripplot(data=tips, x = 'day', y='total_bill',jitter=True, hue='sex')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='black')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='swarm')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='strip')
```

```python
sns.catplot(data=tips, x='sex', kind='count')
```
