```python
import pandas as pd
from sklearn.feature_selection import SelectKBest, chi2
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, 0:-1]
Y = df.iloc[:, -1]

rank = SelectKBest(score_func=chi2, k=5)
X_new = rank.fit(X, Y)
X_new
```

```python
print(X_new.scores_)
```

```python
newSelected = pd.DataFrame(X_new.scores_)
newSelected
```

```python
Xcolumns = pd.DataFrame(X.columns)
Xcolumns
```

```python
selectedtop = pd.concat([Xcolumns, newSelected], axis = 1)
selectedtop.columns = ['Features', 'Scores']
selectedtop
```

```python
selectedtop.nlargest(5, 'Scores')
```

```python
from sklearn.ensemble import ExtraTreesClassifier
```

```python
rank = ExtraTreesClassifier()
new_X = rank.fit(X, Y)
new_X
```

```python
print(new_X.feature_importances_)
```

```python
rankFeatdf = pd.DataFrame(new_X.feature_importances_)
rankFeatdf
```

```python
newrankFeatdf = pd.concat([Xcolumns, rankFeatdf], axis=1)
newrankFeatdf.columns = ['Features', 'Scores']
newrankFeatdf
```

```python
newrankFeatdf.nlargest(5, 'Scores')
```
